```python
import pandas as pd

df = pd.read_csv('/content/kc_house_data.csv', sep=',')
df.head()
```

###1.2 Создать новый признак delta_renovated, который будет содержать разницу в годах между годом реновацией дома и годом постройки дома.Если реновации дома не было, то в новом признаке поставьте 0

```python
difference_in_years = []

for i, row in df.iterrows():
    if row['yr_renovated']:
        difference_in_years.append(row['yr_renovated']-row['yr_built'])
    else:
        difference_in_years.append(0)

df['delta_renovated'] = difference_in_years
df[['yr_built', 'yr_renovated', 'delta_renovated']].head()
```

###1.3 Создайте признаки года продажи, месяца продажи

```python
df['year_of_sale'] = df['date'].apply(lambda x:  int(x[:4]))
df['month_of_sale'] = df['date'].apply(lambda x:  int(x[4:6]))
df[['date', 'year_of_sale', 'month_of_sale']].head()
```

###1.4 Удалите признаки date, zipcode, lat, long

```python
df.drop(columns=['date', 'zipcode', 'lat', 'long'], inplace=True)
df.head()
```
